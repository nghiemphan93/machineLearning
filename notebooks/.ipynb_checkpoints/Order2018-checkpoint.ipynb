```python
kunde415693980 = order2018[order2018['kunden_oid'] == 415693980]
kunde415693980 = kunde415693980.drop_duplicates(subset='liefer_dtm')
kunde415693980
```

```python
kunde415693980NotZero = kunde415693980[kunde415693980['auftragswert_btr_eur'] > 0]
kunde415693980NotZero
```

```python
kunde415693980['dateDifference'] = kunde415693980NotZero['liefer_dtm'].diff().dt.days
kunde415693980['dateAllDifference'] = kunde415693980['liefer_dtm'].diff().dt.days
kunde415693980
```

```python
kunde415693980 = kunde415693980[['kunden_oid', 'liefer_id', 'liefer_dtm','auftragswert_btr_eur', 'dateDifference', 'dateAllDifference']]
```

```python
indexNotNull = list(kunde415693980['dateDifference'].dropna().index)
indexNotNull
```

```python
window = 2
dfLength = len(kunde415693980['dateDifference'].dropna())
for i in range(dfLength - window):
#     print(kunde415693980['dateDifference'].dropna().iloc[i:i+window, ].mean())
#     print(kunde415693980.iloc[i+window])
    print(f'{indexNotNull[i+window]} {kunde415693980["dateDifference"].dropna().loc[indexNotNull[i]:indexNotNull[i+window]].mean().round(2)}')
    kunde415693980.loc[indexNotNull[i+window], 'meanWindow2'] = kunde415693980["dateDifference"].dropna().loc[indexNotNull[i]:indexNotNull[i+window]].mean().round(2)
    kunde415693980.loc[indexNotNull[i+window], 'stdWindow2'] = kunde415693980["dateDifference"].dropna().loc[indexNotNull[i]:indexNotNull[i+window]].std().round(2)
```
